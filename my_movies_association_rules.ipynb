# apriori algorithm

```python
frequent_itemsets=apriori(movie, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
movie[movie["Sixth Sense"]==1].shape[0]/movie.shape[0]
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1.2)
rules
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1.4)
rules
```

```python
rules=association_rules(frequent_itemsets,metric="support",min_threshold=0.3)
rules
```

```python
rules=association_rules(frequent_itemsets,metric="support",min_threshold=0.5)
rules
```

```python
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.5)
rules
```
